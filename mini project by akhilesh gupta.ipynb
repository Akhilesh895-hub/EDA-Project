```python
x = df['price']
y1 = df['area']
plt.figure(figsize=(13,5))
plt.title('HOUSING:-price vs area')
plt.xlabel('price')
plt.ylabel('area')
plt.plot(x,y1, color='r')
plt.show()
```

```python
x = df['price']
y2 = df['bedrooms']
plt.figure(figsize=(12,5))
plt.title('HOUSING:-price vs bedrooms')
plt.xlabel('price')
plt.ylabel('bedrooms')
plt.plot(x,y2, color='k', label='price vs bedroom')
plt.legend()
plt.show()
```

```python
x = df['price']
y2 = df['bedrooms']
y3 = df['bathrooms']
y4 = df['stories']

plt.figure(figsize=(14,6))
plt.title('HOUSING:-price vs bedrooms, bathrooms & stories')
plt.xlabel('price')
plt.xticks(np.arange(1,8,1))
plt.yticks(np.arange(1,8,1))
plt.ylabel('bedrooms, bathrooms & stories')
plt.plot(x,y2, color='k')
plt.plot(x,y3, color='r')
plt.plot(x,y4, color='c')
plt.grid()
plt.show()
```

```python
x = df['price']
y5 = df['guestroom']
y6 = df['airconditioning']
y7 = df['basement']
y8 = df['parking']
y9 = df['prefarea']

plt.figure(figsize=(14,6))
plt.title('HOUSING:-price vs facilities')
plt.xlabel('price')
plt.ylabel('guestroom, airconditioning, basement, parking & prefarea')
plt.plot(x,y5, color='k')
plt.plot(x,y6, color='r')
plt.plot(x,y7, color='c')
plt.plot(x,y8, color='y')
plt.plot(x,y9, color='g')
plt.grid()
plt.show()
```

```python
x = df['price']
y1 = df['area']
y2 = df['bedrooms']
y3 = df['bathrooms']
y4 = df['stories']
y5 = df['guestroom']
y6 = df['airconditioning']
y7 = df['basement']
y8 = df['parking']
y9 = df['prefarea']

plt.figure(figsize=(14,6))
plt.xlabel('price')
plt.ylabel('accomodities')
plt.xticks(np.arange(1,8,1))
plt.yticks(np.arange(1,8,1))


plt.scatter(x,y1, color='orange')
plt.scatter(x,y2, color='k')
plt.scatter(x,y3, color='hotpink')
plt.scatter(x,y4, color='violet')
plt.scatter(x,y5, color='c')
plt.scatter(x,y6, color='pink')
plt.scatter(x,y7, color='r')
plt.scatter(x,y8, color='y')
plt.scatter(x,y9, color='g')

plt.legend(['area', 'bedrooms', 'bathrooms', 'stories', 'guestroom', 'airconditioning', 'basement', 'parking', 'prefarea'])
plt.show()
```
